#### 요약

```python
#요약, 제목 짓기에 활용?
bullet_summ = Pororo(task="text_summarization", lang="ko", model="bullet")
```

```python
text ='''손흥민은 토트넘의 첫 시즌인 2015~2016시즌 EPL에서 28경기에 출전 4골에 그쳤지만 2016~2017시즌부터 주전으로 도약했다.
매 시즌 두 자릿수 골을 터트렸고, 지난 시즌에는 23골을 기록해 아시아 선수 최초로 골든부트(득점왕)를 거머쥐었다.
하지만 손흥민은 활약과는 달리 상대적으로 이적설이 많지 않았다. 유럽이적시장 전문가인 파브리지오 로마노가 26일(현지시각) 자신의 '뉴스레터'를 통해 그 해답을 내놓았다.'''

bullet_summ(
    text, 
    beam=5, 
    len_penalty=0.6,
    no_repeat_ngram_size=3,
    top_k=20,
    top_p=0.8
)
```

```python
text = '사람 마음의 바탕은 선도 악도 아니다. 선과 악은 인연에 따라 일어날 뿐. 선한 인연을 만나면 마음이 선해지고 나쁜 인연을 만나면 마음이 악해진다. 안개 속에 있으면 자신도 모르게 옷이 젖듯이.'


bullet_summ(
    text, 
    beam=5, 
    len_penalty=0.6,
    no_repeat_ngram_size=3,
    top_k=20,
    top_p=0.8
)

```
